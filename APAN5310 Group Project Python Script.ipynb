```python
# agent_customer
import pandas as pd
import psycopg2
from psycopg2 import sql, errors

# Convert all relevant columns to Python int
agent_customer_df['employee_id'] = agent_customer_df['employee_id'].astype(int)
agent_customer_df['customer_id'] = agent_customer_df['customer_id'].astype(int)
agent_customer_df['interested_listing_id'] = agent_customer_df['interested_listing_id'].astype(int)


# Insert data into the agent_customer table with duplicate and foreign key violation handling
for index, row in agent_customer_df.iterrows():
    try:
        delete_query = """
        DELETE FROM agent_customer WHERE employee_id = %s AND customer_id = %s AND interested_listing_id = %s
        """
        cur.execute(delete_query, (int(row['employee_id']), int(row['customer_id']), int(row['interested_listing_id'])))
        
        insert_query = """
        INSERT INTO agent_customer (employee_id, customer_id, interested_listing_id)
        VALUES (%s, %s, %s)
        """
        cur.execute(insert_query, (
            int(row['employee_id']), int(row['customer_id']), int(row['interested_listing_id'])
        ))

        conn.commit()  # Commit the transaction if no errors occur

    except errors.ForeignKeyViolation as e:
        print(f"ForeignKeyViolation: {e}")
        conn.rollback()  # Rollback the transaction in case of error to maintain the database state

    except Exception as e:
        print(f"Error: {e}")
        conn.rollback()  # Rollback the transaction in case of error to maintain the database state
```

```python
# office
# Convert all relevant columns to Python int
office_df['office_id'] = office_df['office_id'].astype(int)


# Insert data into the office table with duplicate handling and foreign key check
for index, row in office_df.iterrows():
    try:
        # Check if the office_id is referenced in other tables
        reference_check_query = """
        SELECT COUNT(*) FROM operational_cost WHERE office_id = %s
        """
        cur.execute(reference_check_query, (int(row['office_id']),))
        reference_count = cur.fetchone()[0]
        
        if reference_count == 0:
            delete_query = """
            DELETE FROM offices WHERE office_id = %s
            """
            cur.execute(delete_query, (int(row['office_id']),))
        
        insert_query = """
        INSERT INTO offices (office_id, state, address)
        VALUES (%s, %s, %s)
        """
        cur.execute(insert_query, (
            int(row['office_id']), row['state'], row['address']
        ))

        conn.commit()  # Commit the transaction if no errors occur

    except errors.ForeignKeyViolation as e:
        print(f"ForeignKeyViolation: {e}")
        conn.rollback()  # Rollback the transaction in case of error to maintain the database state

    except Exception as e:
        print(f"Error: {e}")
        conn.rollback()  # Rollback the transaction in case of error to maintain the database state
```
